# 比较不同的模型性能——超参数(num_layer)

```python
from utils.utils import load_json
from matplotlib import pyplot as plt
from glob import glob
from os.path import join
```

```python
ans = {}
```

```python
for base_dir in glob('result/debug-0527-[1246]'):
    num_layer = load_json(join(base_dir, 'args.json'))['num_layer']
    record = load_json(join(base_dir, 'record.json'))
    ans[num_layer] = record
```

```python
metrics = ['mse','r2']
modes = ['train', 'eval']
```

```python
i = 1
fontsize = 16
plt.figure(figsize=(15,10))
for metric in metrics:
    for mode in modes:
        plt.subplot(2,2,i)
        i += 1
        for num_layer in ans.keys():
            plt.plot(ans[num_layer][metric][mode],label=f'num_layer={num_layer}')
        plt.xlabel('epoch',fontsize=fontsize)
        plt.ylabel(mode + '-' +  metric.upper(),fontsize=fontsize)
        plt.legend(fontsize=fontsize)
        plt.xticks(fontsize=fontsize)
        plt.yticks(fontsize=fontsize)
        plt.grid(which='both')
```
