```python
from utils.utils import load_json, mkdir
import numpy as np
from matplotlib import pyplot as plt
from os.path import join
```

```python
record = load_json('result/debug-0527-1/record.json')
```

```python
plt.figure()
fontsize = 22
figsize = (15/1.5,10/1.5)
fig_dir = 'tmp'
mkdir(fig_dir)
plt.figure(figsize=figsize)
plt.plot(record['mse']['train'],c='r',label='train')
plt.plot(record['mse']['eval'],c='b',label='eval')
plt.legend(loc='upper right',fontsize=fontsize)
plt.xlabel('epoch',fontsize=fontsize)
plt.ylabel('MSE',fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.grid(which='both')
plt.savefig(join(fig_dir, 'mse.png'))
# plt.close()

plt.figure(figsize=figsize)
plt.plot(record['r2']['train'],c='r',label='train')
plt.plot(record['r2']['eval'],c='b',label='eval')
plt.legend(loc='lower right',fontsize=fontsize)
plt.xlabel('epoch',fontsize=fontsize)
plt.ylabel('R2',fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.grid(which='both')
plt.savefig(join(fig_dir, 'r2.png'))
    # plt.close()
```
